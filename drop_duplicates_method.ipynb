```python
import numpy as np
import pandas as pd
```

```python
data={
    'A':[8,11,8,9,16,3,12],
    'B':[5,21,5,18,8,18,51],
    'C':[7,30,7,27,95,30,42]
}
df=pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
df.duplicated(subset=['C'])
```

```python
df
```

```python
df.duplicated(subset=['C'])
```

```python
df.duplicated(subset=['C'], keep='last')
```

```python
df.duplicated(subset=['C'], keep=False)
```

```python
df.duplicated(keep='last')
```

```python
df.drop_duplicates()
```
