```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
# Example 1: Filling NaN values with a scalar
df1 = pd.DataFrame({'A': [1, 2, None, 4]})
#print("Before:\n", df1)
df1.fillna(0, inplace=True)
#print("\nAfter:\n", df1)
```

```python
df1 = pd.DataFrame({'A': [1, 2, None, 4]})
df1.fillna(0, inplace=True)
```

```python
ddd={'Name':['John','Alice','Bob','Usher',np.nan,'Ali'],
     'Age':[25,np.nan,38,np.nan,40,21],
     'Salary':[np.nan,15000,28000,31000,40000,np.nan]
     }
df=pd.DataFrame(ddd)
df
```

```python
df.fillna(-1, axis=0)
```

```python
df['Age'].mean()
```

```python
df.fillna({'Age':31, 'Salary':0,'Name':'Sally'})
```

```python
df.fillna({'Age':df['Age'].mean(),'Salary':df['Salary'].mean()})
# bu yerda biz har bir column uchun o'rtacha qiymatni topib
# so'ng o'sha qatordagi NaN qiymatarni to'ldiryapmiz
```
