```python
import numpy as np
import pandas as pd
```

```python
dict1 = {
    'January': [17, 5.83, -4, 3.14, 10, -0.07, 6.78,12,-5],
    'February': [8, 2.45, -3, 6.78, 15, -0.12, 22, 9.9, -7]
}

ind = [['USA','USA','USA','Canada','Canada','Canada','Mexico','Mexico','Mexico'],
["New York", "Los Angeles", "Chicago", "Toronto", "Vancouver", "Montreal", "Mexico City", "Guadalajara", "Monterrey"]]
df = pd.DataFrame(dict1, index=ind)
df
```

```python
df.loc[ [('Canada', 'Vancouver'), ('Canada', 'Toronto') ]]
df.loc[[('Canada','Montreal'),('Mexico','Monterrey'),('USA','Chicago')]]
```

```python
df.loc[('USA','Chicago')]
```

```python
df.sort_index().loc['USA':'Canada']
```

```python
df_sorted = df.sort_index()

# Accessing data for 'USA' and 'Canada' only
values_usa_canada = df_sorted.loc['USA':'Canada']
print(values_usa_canada)
```

```python
# Accessing data for 'USA' and 'Canada' only
values_usa_canada = df[(df.index.get_level_values(0) == 'USA') | (df.index.get_level_values(0) == 'Canada')]
print(values_usa_canada)
```
