```python
import pandas as pd
import numpy as np
```

```python
data={
    'Salary':np.random.randint(1000,100000,15),
    'Name':['James','John','Robert','Michael','William','David','Richard','Charles','Joseph','Thomas','Christopher','Daniel','Paul','Mark','Donald']
    }
df=pd.DataFrame(data)
df
```

```python
data={'ages' : [22, 35, 47, 50, 25, 34, 40, 55]}
bins = [20, 35, 50, 60]
labels=['Low','Medium','High']
df2=pd.DataFrame(data)
guruhlar=pd.cut(df2['ages'], bins=bins)
guruhlar
```

```python
guruhlar.value_counts()
```

```python
df2['intervals']=pd.cut(df2['ages'], bins=bins)
df2
```

```python
status=['Low','Medium','High']
pd.cut(df2['ages'], bins=bins, labels=status)
```

```python
df2['status']=pd.cut(df2['ages'], bins=bins, labels=status)
df2
```

```python
a=pd.cut(df2['ages'], bins=bins, labels=status)
a.value_counts()
```

```python
df2['intervals']=pd.cut(df2['ages'], bins=3)
df2
```

```python
range=pd.cut(df2['ages'], bins=3)
range.value_counts()
```
