```python
import numpy as np
import pandas as pd
#groupby yordamida guruhlash
```

```python
data = {
    'Category': ['A', 'A', 'B', 'B', 'A'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'Z'],
    'Sales': [100, 200, 150, 250, 120],
    'Price': [10, 20, 15, 25, 12],
    'Country': ['USA', 'Canada', 'USA', 'Canada', 'Mexico'],
    'Revenue': [1000, 2000, 1500, 2500, 1200],
    'Size': ['S', 'M', 'M', 'L', 'S'],
    'Region': ['East', 'West', 'East', 'West', 'East'],
    'CustomerID': [1, 2, 3, 4, 5],}
df = pd.DataFrame(data)
df
```

```python
# 1-misol 1ta column bo'yicha guruhlash
df.groupby('Category')['Price'].sum()
```

```python
# 2-misol 2ta column bo'yicha guruhlash
df.groupby(['Category','Region'])['Price'].sum()
```

```python
# 3-misol
df.groupby('Category',group_keys=False)['Sales'].apply(lambda x : x/2)
```

```python
df.groupby('Country')['Sales'].mean()
```

```python
df.groupby('Country')['Sales'].agg(['mean','sum'])
```

```python
df['Sales'].agg(['mean','sum'])
```

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}
df = pd.DataFrame(data)
df
```

```python
df.agg('mean')
```

```python
df.apply({'A':'mean','B':'sum','C':'max'})
```

```python
dict1={
    'Fruit':['Apple','Lemon','Orange','Lemon','Apple','Orange'],
    'Price':[2500,3200,5700,1900,3300,4100]
}
df=pd.DataFrame(dict1)
df
```

```python
df.groupby('Fruit')['Price'].agg(['max','mean','sum'])
```

```python
df['new']=df.groupby('Fruit')['Price'].transform(lambda x : x.max()-x.min())
df
```
