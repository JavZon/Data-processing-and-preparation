```python
url='https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data_nan.csv'
df=pd.read_csv(url, index_col=0)
df.head()
```

```python
df['engine-type'].value_counts()
```

```python
dict1={'dohc':'dual overhead cam','ohc':'overhead cam',
       'ohcv':'overhead cam veh', 'ohcf':'overhead cam veh function',
       'I':'inline engine','dohcv':'dual overhead cam veh','NaN':0}
df['new_eng']=df['engine-type'].map(dict1)
df
```

```python
data = {'score': [75, 90, 60, 85, 40]}
df = pd.DataFrame(data)
df
```

```python
df['category']=df['score'].map(lambda x : 'High' if x >80 else 'Medium' if 65 < x < 80  else 'Low')
df
```

```python
def fahrenheit_to_celsius(temp):
    return (temp - 32) * 5/9

data = {'temp_f': [34, 68, 86, 104]}
df = pd.DataFrame(data)
df
```

```python
df['temp_celsius']=df['temp_f'].map(lambda f : (f-32)*5/9)
df
```

```python
df['temp_celsius']=df['temp_f'].apply(lambda f : (f-32)*5/9)
df
```

```python
dict2={
    'Name':['Anvar','Ali','John','Vali','Nick'],
    'Category':[15,20,21,33,41],
    'Quantity':['four','two','three','five','ten'],
    'Age':[38,48,51,15,20]
}
df=pd.DataFrame(dict2)
df
```
