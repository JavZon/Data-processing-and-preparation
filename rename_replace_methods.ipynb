```python
import pandas as pd
import numpy as np
```

```python
data={
    'Club':['Arsenal','Liverpool','Man City','Aston Villa','Tottenham'],
    'MP':[28,28,28,28,27],
    'Win':[20,19,19,17,16],
    'Draw':[4,7,6,4,5],
    'Loss':[4,2,3,7,6],
    'Points':[64,64,63,55,53]
}
df=pd.DataFrame(data)
df
```

```python
df.replace({3:99,2:100})
```

```python
data2={
    'A':['John','Ali','Wick','Olim','Anvar'],
    'B':['four','five','six','three', 'two'],
    'C':[25,32,48,17,23]
}
df=pd.DataFrame(data2)
df
```

```python
df.replace(to_replace={'four':4, 'five':5, 'six':6, 'three':3, 'two':2})
df
```

```python
df
```

```python
df.rename(index={0:'r1',1:'r2',2:'r3',3:'r4',4:'r5'}, columns={'A':'Alpha'}, inplace=True)
df
```

```python
df.rename(index=str.upper, columns=str.upper)
```

```python
df.rename(columns={'C':'Calvin','B':'Betta'})
```
